## Librerias

```python
## Importacion de librearias para es desarrollo de la actividad

import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import seaborn as sns
```

## Datos que se van a manipular

```python
## Creacion de variable con el proposito de listar sobre filtro datos quye conetengan el valor de la cadena
data_name=[x for x in os.listdir() if 'Data_rndc_year_2024.csv' in x][0]
```

```python
## Lectrua del archivo y codificacion adicional traemos los primero 1000 datos
df_rndc = pd.read_csv(data_name, sep=';', encoding='utf-8').head(1000)
df_rndc
```

## Analisis de columnas

```python
## Identificacion de columnas 
df_rndc.columns
```

```python
## Filtrado de columnas potenciales
df_rndc_columns=df_rndc[["Item","Fecha","Sucursal","Modalidad","Peso","Origen","Destino","Valor Orden","Costo Orden","Rentabilidad","Flete Factura","Factura"]]
df_rndc_columns
```

```python
##
df_rndc_sum=df_rndc_columns.value_counts()
df_rndc_sum
```

## Analisis exploratorio de datos EDA (Exploratory Data Analysis)

```python
a1=df_rndc_columns[['Sucursal','Fecha','Item']]
sns.boxplot(data=a1, x='Sucursal', y='Item', showfliers=False)
plt.xlabel('Sucursal')
plt.ylabel('Item')
plt.title('Boxplot manifiestos de carga')
```

```python
a1=df_rndc_columns[['Sucursal','Item','Modalidad']]
sns.boxplot(data=a1, x='Sucursal', y='Item',hue='Modalidad',showfliers=False)
plt.xlabel('Sucursal')
plt.ylabel('Items')
plt.title('Boxplot manifiestos de carga')
```

```python
sns.boxplot(data=a1, x='Sucursal', y='Item', showfliers=True)
```

```python
sns.histplot(a1['Modalidad'])
```

```python
x=df_rndc['Modalidad'].value_counts().index
y=df_rndc['Modalidad'].value_counts().values
sns.barplot(x=x, y=y)
plt.xticks(rotation=45)
plt.xlabel('Modalidad')
plt.ylabel('Item')
plt.title('Reportes por modalidad')
```
