## Analisis exploratorio de datos EDA (Exploratory Data Analysis)

```python
a1=df_rndc_columns[['Sucursal','Fecha','Item']]
sns.boxplot(data=a1, x='Sucursal', y='Item', showfliers=False)
plt.xlabel('Sucursal')
plt.ylabel('Item')
plt.title('Boxplot manifiestos de carga')
```

```python
a1=df_rndc_columns[['Sucursal','Fecha','Modalidad']]
sns.boxplot(data=a1, x='Sucursal', y='Fecha',hue='Modalidad',showfliers=False)
plt.xlabel('Sucursal')
plt.ylabel('Fecha')
plt.title('Boxplot manifiestos de carga')
```

```python
sns.boxplot(data=a1, x='Sucursal', y='Fecha', showfliers=True)
```

```python
sns.histplot(a1['Modalidad'],bins=3)
```

```python
x=df_rndc['Modalidad'].value_counts().index
y=df_rndc['Modalidad'].value_counts().values
sns.barplot(x=x, y=y)
plt.xticks(rotation=45)
plt.xlabel('Modalidad')
plt.ylabel('Fecha')
plt.title('Reportes por modalidad')
```
